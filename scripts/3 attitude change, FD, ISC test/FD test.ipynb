```python

# /Volumes/Li/task-persuade/fmriprep/sub-01/func/sub-01_task-persuade_run-1_desc-confounds_timeseries.tsv
# /Volumes/Li/task-persuade/fmriprep/sub-013/func/sub-013_task-persuade_run-1_desc-confounds_timeseries.tsv

import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sub_list = [f'sub-{x:0>2d}' for x in range(1,36)]
run_list = list(range(1,6))
```

```python
subs_fd = []
fd_dict = {}
for sub in sub_list:
    fds = []
    for run in run_list:
        confound_file = f'/Volumes/Li/task-persuade/fmriprep/{sub}/func/{sub}_task-persuade_run-{run}_desc-confounds_timeseries.tsv'
        confound_df = pd.read_csv(confound_file, delimiter ='\t')
        fd = confound_df['framewise_displacement']
        fds.append(fd)
        
    fd_all = pd.concat(fds, ignore_index=True)
    fd_dict[sub] = fd_all
    
    fds.append(fd_all.mean())
    
    if fd_all.mean() > 0.5:
        print(sub,fd_all.mean())
    subs_fd.append(fd_all.mean())
np.mean(subs_fd)         
```

```python
fd_df = pd.DataFrame(fd_dict)
fd_df.head()
```

```python
fd_df = pd.DataFrame(fd_dict)

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 4))
ax = sns.boxplot(data=fd_df, color='grey', showfliers=False)

ax.set_xticklabels([f'sub-{i+1}' for i in range(len(ax.get_xticklabels()))])

# 添加标题和坐标轴标签
plt.axhline(0.5, color='black', linestyle='--', label='Threshold = 0.5')
plt.xticks(rotation=45)
plt.xlabel('Subject')
plt.ylabel('Framewise Displacement (FD)')
plt.title('FD Distribution per Subject in Study 1')
plt.legend()

plt.gca().set_facecolor('white')  # 背景设置为白色
plt.gca().spines['top'].set_visible(False)  # 去掉上框线
plt.gca().spines['right'].set_visible(False)  # 去掉右框线
plt.gca().spines['left'].set_linewidth(1.5)  # 设置左框线宽度
plt.gca().spines['bottom'].set_linewidth(1.5)  # 设置下框线宽度
plt.gca().grid(False)  # 关闭网格线

plt.tight_layout()
plt.savefig(f"/Users/li/Desktop/debate2025/Figures/fd_per_sub_persuade.png", dpi=300, bbox_inches='tight')

plt.show()
```
