```python
import numpy as np
import pandas as pd
```

```python
excel_file = '/Users/li/Desktop/ratedata/questionnaire.xlsx'
data = pd.read_excel(excel_file)
print(data.columns)
data.head()
```

```python
# AP = Approach Subscale, AV = Avoidance Subscale
# To build an aggregate score of the Need for Affect, avoidance items must be reverse scored.
NFA_AV = ['nfa1', 'nfa4', 'nfa6', 'nfa7', 'nfa10']
NFA_AP = ['nfa2', 'nfa3', 'nfa5', 'nfa8', 'nfa9']

# 'nfc3', 'nfc4' is reverse coded items.
NFC = ['nfc1', 'nfc2', 'nfc3', 'nfc4', 'nfc5', 'nfc6']

# 'iri2', 'iri5', 'iri10', 'iri11', 'iri14' is reverse coded items.
IRI_PT = ['iri6','iri9', 'iri15', 'iri19', 'iri22']
IRI_FS = ['iri3', 'iri5', 'iri10', 'iri12', 'iri17', 'iri20']
IRI_EC = ['iri1', 'iri2', 'iri7', 'iri11', 'iri14', 'iri16']
IRI_PD = ['iri4', 'iri8', 'iri13', 'iri18', 'iri21']

# PA = Prospective Anxiety, IA = Inhibitory Anxiety, 
IUS_PA = ['iou1', 'iou2', 'iou3', 'iou4', 'iou5', 'iou6', 'iou7']
IUS_IA = ['iou8', 'iou9', 'iou10', 'iou11', 'iou12']

#
know = ['know1', 'know2', 'know3', 'know4', 'know5', 'know6', 'know7', 'know8']
like = ['like1', 'like2', 'like3', 'like4', 'like5', 'like6', 'like7', 'like8']
post_attitude = ['attd1', 'attd2', 'attd3', 'attd4', 'attd5', 'attd6', 'attd7', 'attd8', 'attd9', 'attd10']

# reverse items
reverse_items = ['nfa1', 'nfa4', 'nfa6', 'nfa7', 'nfa10', 
                'nfc3', 'nfc4',
                'iri2', 'iri5', 'iri10', 'iri11', 'iri14']
```

```python
for item in ['nfa1', 'nfa4', 'nfa6', 'nfa7', 'nfa10']:
    data[item] = 8 - data[item]
for item in ['nfc3', 'nfc4']:
    data[item] = 6 - data[item]
for item in ['iri2', 'iri5', 'iri10', 'iri11', 'iri14']:
    data[item] = 4 - data[item]
```

```python

first_factors = {
    
    # 'sub': ['sub'],
    # 'sex': ['sex'],
    # 'age': ['age'],
    
    'NFA_AV': ['nfa1', 'nfa4', 'nfa6', 'nfa7', 'nfa10'],
    'NFA_AP': ['nfa2', 'nfa3', 'nfa5', 'nfa8', 'nfa9'],

    'NFC': ['nfc1', 'nfc2', 'nfc3', 'nfc4', 'nfc5', 'nfc6'],

    'IRI_PT': ['iri6','iri9', 'iri15', 'iri19', 'iri22'],
    'IRI_FS': ['iri3', 'iri5', 'iri10', 'iri12', 'iri17', 'iri20'],
    'IRI_EC': ['iri1', 'iri2', 'iri7', 'iri11', 'iri14', 'iri16'],
    'IRI_PD': ['iri4', 'iri8', 'iri13', 'iri18', 'iri21'],

    'IUS_PA': ['iou1', 'iou2', 'iou3', 'iou4', 'iou5', 'iou6', 'iou7'],
    'IUS_IA': ['iou8', 'iou9', 'iou10', 'iou11', 'iou12']
}

first_averages = pd.DataFrame()
for dimension, columns in first_factors.items():
    dimension_data = data[columns]
    first_averages[dimension] = dimension_data.sum(axis=1)
    
second_factors = {
    'NFA': first_factors['NFA_AV'] + first_factors['NFA_AP'],
    'IRI': first_factors['IRI_PT'] + first_factors['IRI_FS'] + first_factors['IRI_EC'] + first_factors['IRI_PD'],
    'IUS': first_factors['IUS_PA'] + first_factors['IUS_IA']
}

second_averages = pd.DataFrame()

for dimension, columns in second_factors.items():
    dimension_data = data[columns]
    second_averages[dimension] = dimension_data.sum(axis=1)

# 加上名字
second_averages= pd.concat([data[['sub', 'sex', 'age']], second_averages, first_averages], axis=1)
second_averages.to_excel('/Users/li/Desktop/ratedata/personality.xlsx', index = False)
second_averages.head()
```
