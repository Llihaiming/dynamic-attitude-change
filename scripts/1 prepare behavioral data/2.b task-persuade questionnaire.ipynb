```python
import numpy as np
import pandas as pd

csv_file = '/Users/li/Desktop/task-persuade/behavdata/questionnaire.csv'
data = pd.read_csv(csv_file)
print(data.columns)
data.head()
# sub 1-35
# sex 男0女1
# a1-a10 情感需求
# b1-b6 认知需求
# c1-c28 人际指数
# d1-d12 不确定性
# e1-e60 大五人格
# f1-f12 社会影响易感性
```

```python
# AP = Approach Subscale, AV = Avoidance Subscale
# To build an aggregate score of the Need for Affect, avoidance items must be reverse scored.
NFA_AV = ['a1', 'a4', 'a6', 'a7', 'a10']
NFA_AP = ['a2', 'a3', 'a5', 'a8', 'a9']

# 'nfc3', 'nfc4' is reverse coded items.
NFC = ['b1', 'b2', 'b3', 'b4', 'b5', 'b6']

# 'c2', 'c5', 'c10', 'c11', 'c14' is reverse coded items.
IRI_PT = ['c6','c9', 'c15', 'c19', 'c22']
IRI_FS = ['c3', 'c5', 'c10', 'c12', 'c17', 'c20']
IRI_EC = ['c1', 'c2', 'c7', 'c11', 'c14', 'c16']
IRI_PD = ['c4', 'c8', 'c13', 'c18', 'c21']

# PA = Prospective Anxiety, IA = Inhibitory Anxiety, 
IUS_PA = ['d1', 'd2', 'd3', 'd4', 'd5', 'd6', 'd7']
IUS_IA = ['d8', 'd9', 'd10', 'd11', 'd12']

Big5 = [f'e{x}' for x in range(1,61)]

SIS = ['f1','f2','f3','f4','f5','f6','f7','f8','f9','f10','f11','f12']
```

```python

# reverse items
reverse_items = ['a1', 'a4', 'a6', 'a7', 'a10', 
                'b3', 'b4',
                'c2', 'c5', 'c10', 'c11', 'c14']

for item in ['a1', 'a4', 'a6', 'a7', 'a10']:
    data[item] = 8 - data[item]
for item in ['b3', 'b4']:
    data[item] = 6 - data[item]
for item in ['c2', 'c5', 'c10', 'c11', 'c14']:
    data[item] = 4 - data[item]
    
```

```python

first_factors = {
    
    'NFA_AV': ['a1', 'a4', 'a6', 'a7', 'a10'],
    'NFA_AP': ['a2', 'a3', 'a5', 'a8', 'a9'],

    'NFC': ['b1', 'b2', 'b3', 'b4', 'b5', 'b6'],

    'IRI_PT': ['c6','c9', 'c15', 'c19', 'c22'],
    'IRI_FS': ['c3', 'c5', 'c10', 'c12', 'c17', 'c20'],
    'IRI_EC': ['c1', 'c2', 'c7', 'c11', 'c14', 'c16'],
    'IRI_PD': ['c4', 'c8', 'c13', 'c18', 'c21'],

    'IUS_PA': ['d1', 'd2', 'd3', 'd4', 'd5', 'd6', 'd7'],
    'IUS_IA': ['d8', 'd9', 'd10', 'd11', 'd12'],

    'SIS': ['f1','f2','f3','f4','f5','f6','f7','f8','f9','f10','f11','f12']
}

first_averages = pd.DataFrame()
for dimension, columns in first_factors.items():
    dimension_data = data[columns]
    first_averages[dimension] = dimension_data.sum(axis=1)
    
second_factors = {
    
    'NFA': first_factors['NFA_AV'] + first_factors['NFA_AP'],
    'NFC': first_factors['NFC'],
    'IRI': first_factors['IRI_PT'] + first_factors['IRI_FS'] + first_factors['IRI_EC'] + first_factors['IRI_PD'],
    'IUS': first_factors['IUS_PA'] + first_factors['IUS_IA'],
    'SIS': first_factors['SIS']
}

second_averages = pd.DataFrame()

for dimension, columns in second_factors.items():
    dimension_data = data[columns]
    second_averages[dimension] = dimension_data.sum(axis=1)

# 
second_averages= pd.concat([data[['sub', 'sex', 'age']], second_averages, first_averages], axis=1)
second_averages.to_csv('/Users/li/Desktop/task-persuade/behavdata/personality.csv', index = False)
second_averages.head()
```
